```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#create data frame
data_dict ={'name':['john','David','Peter','jack','mack'],
            'age': [23, 36, 45, 32,44],
            'marks': [87, 91, 95, 88,90]
            }
df = pd.DataFrame(data_dict)
```

```python
#sort score in descending order and reset index
sorted_df = df.sort_values(by='marks',ascending=False)
sorted_df = sorted_df.reset_index(drop=True)
print(sorted_df)
```

```python
#using matplotlib
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.plot(x,y,marker ='o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot')
plt.show()
```

```python
categories =['A','B','C','D']
values = [10,35,50,45]
plt.bar(categories,values,color='white',edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
fig,axis = plt.subplots(2,2)
axis[0,0].plot(x,y,marker='o')
axis[0,1].bar(categories,values,color='black',edgecolor='red')
axis[1,0].hist(values)
axis[1,1].scatter(x,y)
```
