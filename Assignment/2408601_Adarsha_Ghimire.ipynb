```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

Path = '/content/drive/MyDrive/AssignmentAi/WHR-2024-5CS037.csv'
df = pd.read_csv(Path)
df
```

```python
#First ten rows
rows = df.head(10)
print("First Ten Rows: \n",rows)
```

```python
#num of rows and columns
Shape = df.shape
print("\nNumber of rows and columns:\n ", Shape)
```

```python
#datatype and columns
columns = df.columns
print("\n Number of columns: \n", columns)

datatype = df.dtypes
print("Datatype: \n ", datatype)
```

```python
#Mean for Score column
mean = df['score'].mean()
print("mean", mean)
median = df['score'].median()
print("Median: ", median)
#Highest and Lowest
Highest = df['Freedom to make life choices'].max()
print("Highest: ", Highest)
Lowest = df['Freedom to make life choices'].min()
print("Lowest: ", Lowest)
```

```python
#missing values and  count for each column
missingVlaue = df.isnull().sum()
print("Missing Values: \n", missingVlaue)
```

```python
#Filter dataset
Filter_dataset = df[df['score']>7.5]
print("Countries with Score greater than 7.5: \n", Filter_dataset)
```

```python
#sortiing and displaying top 10 rows
sort = df.sort_values(by='Log GDP per capita',ascending=False).head(10)
print("sorted values:\n", sort)
```

```python
#new column
def Happiness_Category(score):
  if(score>6):
    return 'High'
  elif(4<score<6):
    return 'Medium'
  else:
    return 'Low'

df['Happiness_Category'] = df['score'].apply(Happiness_Category)
print(df)
```
